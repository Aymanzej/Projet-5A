```python
import pandas as pd
import os

def export_dataframe_to_csv(df: pd.DataFrame, filename: str) -> str:
    """
    Crée un fichier CSV à partir d'un DataFrame pandas.

    Args:
        df (pd.DataFrame): Le DataFrame à exporter (par exemple, df_moda).
        filename (str): Le nom du fichier CSV souhaité (doit inclure '.csv').

    Returns:
        str: Le chemin d'accès complet au fichier créé.
    """
    # 1. Définir le chemin complet du fichier dans le répertoire actuel
    chemin_complet = os.path.join(os.getcwd(), filename)
    
    # 2. Vérifier et ajouter l'extension .csv si elle manque
    if not chemin_complet.lower().endswith('.csv'):
        chemin_complet += '.csv'

    # 3. Exporter le DataFrame
    try:
        # index=True pour inclure l'index (vos coordonnées time, lat, lon) dans le CSV
        df.to_csv(chemin_complet, index=True) 
        print(f"✅ Export réussi : {chemin_complet}")
        return chemin_complet
    except Exception as e:
        print(f"❌ Erreur lors de l'exportation CSV : {e}")
        return ""
```
