```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
import tensorflow.keras
```

```python
# Création de données synthétiques
x = np.random.uniform(0, 10, 100)
y = np.cos(x)
x_test = np.linspace(2, 8, 100)
y_test = np.cos(x_test)
```

```python
# architecture du modèle
inputs = tf.keras.Input(shape=(1,))
x1 = tf.keras.layers.Dense(64, activation='relu')(inputs)
x2 = tf.keras.layers.Dense(64, activation='relu')(x1)
outputs = tf.keras.layers.Dense(1)(x2)
model = tf.keras.Model(inputs=inputs, outputs=outputs)

optimizer = tf.keras.optimizers.Adam(learning_rate=0.001)
model.compile(optimizer=optimizer, loss='mse')
model.fit(x, y, epochs=300, batch_size=8)
```

```python
y_pred = model.predict(x_test)
```

```python
plt.plot(x_test, y_test, label='Données réelles')
plt.plot(x_test, y_pred, label='Prédictions du modèle')
plt.show()
```

```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
import tensorflow.keras

# Création de données synthétiques
x = np.random.uniform(0, 10, 100)
y = np.cos(x)

x_test = np.linspace(2, 8, 100)
y_test = np.cos(x_test)

# Architecture du modèle
inputs = tf.keras.Input(shape=(1,))
x1 = tf.keras.layers.Dense(64, activation='relu')(inputs)
x2 = tf.keras.layers.Dense(64, activation='relu')(x1)
outputs = tf.keras.layers.Dense(1)(x2)
model = tf.keras.Model(inputs=inputs, outputs=outputs)

optimizer = tf.keras.optimizers.Adam(learning_rate=0.001)
model.compile(optimizer=optimizer, loss='mse')
model.fit(x, y, epochs=300, batch_size=8)

# Prédiction
y_pred = model.predict(x_test)

# -----------------------------
# Affichage du plot
# -----------------------------
plt.figure(figsize=(10, 6))

plt.plot(x_test, y_test, label='Fonction réelle $\\cos(x)$')
plt.plot(x_test, y_pred, label='Prédictions du réseau de neurones')

plt.title("Réseau de neurones en 1D appliqué à la fonction cosinus")
plt.xlabel("Variable d’entrée $x$")
plt.ylabel("Valeur de la fonction $f(x)$")

plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
